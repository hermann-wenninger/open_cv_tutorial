```python
def add(vectors):
    return (sum([v[0] for v  in vectors]), sum([v[1] for v  in vectors]))
```

```python
def translate(translation, vectors):
    return [add(translation, v) for v  in vectors]
```

```python
def draw(d_vectors=d_vectors):
    plt.plot([i[0] for i in d_vectors],[i[1] for i in d_vectors],'ro')
    plt.plot([i[0] for i in d_vectors],[i[1] for i in d_vectors])
    plt.show()

draw(d_vectors)
```
