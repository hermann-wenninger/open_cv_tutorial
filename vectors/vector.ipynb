```python
def add(vectors):
    return (sum([v[0] for v  in vectors]), sum([v[1] for v  in vectors]))
```

```python
def translate(translation, vectors):
    return [add(translation, v) for v  in vectors]
```

```python
def draw(d_vectors=d_vectors):
    plt.plot([i[0] for i in d_vectors],[i[1] for i in d_vectors],'ro')
    plt.plot([i[0] for i in d_vectors],[i[1] for i in d_vectors])
    plt.show()

draw(d_vectors)
```

```python
def add(x,y):
    return x + y

def base_vector_translation(base_vector):
    base_list = []
    translations = [(12*x,10*y) for x in range(-5,5) for y in range(-5,5)]
    for i in range(len(translations)):
        for j in range(len(base_vector)):
            #print(translations[i],base_vector[j])
            x = add(translations[i][0], base_vector[j][0])
            y = add(translations[i][1], base_vector[j][1])
            z = x,y
           
            base_list.append(z)
            #print(base_list)
    draw(base_list)
base_vector_translation(d_vectors)
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, axs = plt.subplots(2, 2, layout='constrained')
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma',
                          norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

pc = axs[1, 1].scatter(data1, data2, c=data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')
```
